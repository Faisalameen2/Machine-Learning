# First ML Model

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
print("Key of iris_dataset:\n{}".format(iris_dataset.keys()))
```

```python
print(iris_dataset["DESCR"][:200] + "\n")
```

```python
print("Target names: \n{}".format(iris_dataset["target_names"]))
```

```python
print("Features names:\n{}".format(iris_dataset["feature_names"]))
```

```python
print("Type of data:\n{}".format(type(iris_dataset["data"])))
```

```python
print("Shape of data:\n{}".format(iris_dataset["data"].shape))
```

```python
print("first five columns of data\n{}".format(iris_dataset["data"][0:5]))
```

```python
print("Shape of target{}".format(iris_dataset["target"].shape))
```

```python
 print("Type of Targer:{}".format(iris_dataset["target"]))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset["data"],
                                                   iris_dataset["target"])#, 
                                                   #random_state=0)
```

```python
print("X_train shape", X_train.shape)
print("y_train shape", y_train.shape)
```

```python
print("X_test shape", X_test.shape)
print("y_test shape", y_test.shape)
```

```python
from pandas.plotting import scatter_matrix
import mglearn
```

```python
iris_df = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
grr = scatter_matrix(iris_df, c = y_train, figsize = (15,15), marker = "o",
                       hist_kwds={"bins": 20}, s = 60, alpha = 0.8, 
                       cmap = mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train, y_train)
```

```python
X_new = np.array([[5,2.9, 1, 0.2]])
print("New shape", X_new.shape)
```

```python
prediction = knn.predict(X_new)
print("Predection:", prediction)
print("predection target name", iris_dataset["target_names"][prediction])
```

```python
y_pred = knn.predict(X_test)
print(y_pred)
```

```python
print("Test set score: {:.2f}".format(np.mean(y_pred == y_test)))
```

```python
print("the score = ", knn.score(X_test, y_test))
```
